```python
## Task 1: Mark all launch sites on a map
# Creamos el mapa centrado en el centro aproximado de todos los sitios
site_map = folium.Map(location=[28.5, -80.6], zoom_start=5)

# Añadimos los círculos y etiquetas para cada sitio
for index, row in launch_sites_df.iterrows():
    location = [row['Lat'], row['Long']]
    site_name = row['Launch Site']
    
    # Círculo
    circle = folium.Circle(
        location,
        radius=1000,
        color='#007849',
        fill=True,
        fill_opacity=0.4
    ).add_child(folium.Popup(site_name))
    
    # Texto (etiqueta)
    marker = folium.Marker(
        location,
        icon=DivIcon(
            icon_size=(150,36),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12px; color: #007849;"><b>{site_name}</b></div>',
        )
    )
    
    # Añadir al mapa
    site_map.add_child(circle)
    site_map.add_child(marker)

# Mostrar el mapa (en entorno Jupyter compatible con Folium)
site_map
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
